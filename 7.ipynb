```python
#encoded 
print(df.columns)
df_encoded=pd.get_dummies(
    df,
    columns=["Weather_Conditions","Traffic_Conditions","Order_Time"],
    dtype=int
)
print(df_encoded.columns)
```

```python
# Step 3 - Feature Engineering

print(df_encoded.columns)
df_encoded['Rush_Hour'] = np.where(
    (df_encoded['Order_Time_Morning'] == 1) | (df_encoded['Order_Time_Evening'] == 1),
    'Rush_Hour',
    'Non_Rush_Hour'
)
print(df_encoded["Rush_Hour"])


```

```python
# Convert Delivery_Time into a binary label (1 = Delayed, 0 = Fast)
# Adjust the threshold based on your dataset — here we assume 30 minutes as example
median=df_encoded["Delivery_Time"].median()
print(median)
df_encoded['Delivery_Status'] = data['Delivery_Time'].apply(lambda x: 1 if x > median else 0) #1=delay 0=fast
```

```python
print(df_encoded.columns)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

os.makedirs("data/Fast", exist_ok=True)
os.makedirs("data/Delayed", exist_ok=True)

features = df_encoded.drop(['Order_ID','Customer_Location','Restaurant_Location'], axis=1)

# convert non-numeric columns to numeric (if any)
features = features.apply(pd.to_numeric, errors='coerce')

labels = df_encoded['Delivery_Status']

for i, row in enumerate(features.values):
    img = row.reshape((2, 11))  # reshape numeric features into a small 2D grid
    plt.imsave(f"data/{'Delayed' if labels[i]==1 else 'Fast'}/{i}.png", img, cmap='viridis')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

def create_dense_model():
    model = Sequential([
        Dense(64, activation='relu', input_shape=(features.shape[1],)),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')  # binary classification
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
from sklearn.model_selection import KFold
import numpy as np

X = np.array(features)
y = np.array(labels)

kf = KFold(n_splits=5, shuffle=True, random_state=42)
acc = []

for train_idx, val_idx in kf.split(X):
    model = create_dense_model()
    model.fit(X[train_idx], y[train_idx], epochs=5, verbose=0)
    _, accuracy = model.evaluate(X[val_idx], y[val_idx], verbose=0)
    acc.append(accuracy)

print("Average Accuracy:", np.mean(acc))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = (model.predict(X) > 0.5).astype("int32")
print(confusion_matrix(y, y_pred))
print(classification_report(y, y_pred))
```
